```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import boto3
import io
import os
import PIL
import tensorflow as tf
```

```python
s3 = boto3.resource('s3')
s3_client = boto3.client('s3')
bucket_name = 'yugiohcardimages'
bucket = s3.Bucket(bucket_name)
bucket.objects
```

```python
bucket.objects
```

```python
object_iterator = bucket.objects.all()
images = []
```

```python
for idx, obj in enumerate(object_iterator):
    full_object = s3_client.get_object(Bucket=bucket_name, Key=obj.key)
    images.append(PIL.Image.open(io.BytesIO(full_object['Body'].read())))
    if idx > 10:
        break
```

```python
width, height = images[0].size
width, height
```

```python
sizing_image_dir = 'sizing'
sizing_image_keys = [sizing_image_dir + '/Left.jpg', sizing_image_dir + '/Top.jpg', 
                     sizing_image_dir + '/Right.jpg', sizing_image_dir + '/Bottom.jpg']
sizing_images = []
for img_key in sizing_image_keys:
    img_object = s3_client.get_object(Bucket=bucket_name, Key=img_key)
    sizing_images.append(PIL.Image.open(io.BytesIO(img_object['Body'].read())))
```

```python
original_width, original_height = images[0].size

left_pixel = original_width - sizing_images[0].width
top_pixel = original_height - sizing_images[1].height
right_pixel = sizing_images[2].width
bottom_pixel = sizing_images[3].height
```

```python
images[0].crop([left_pixel, top_pixel, right_pixel, bottom_pixel])
```
